```python
#DATA CLEANING
#percent of null values
percent_null = df.notnull().sum()*100/df.shape[0]
percent_null
```

```python
#to put missin/null value as unknown in df
df.fillna(value = 'unknown',inplace = True)
print(df.isnull().sum())

```

```python
#to drop unwanted columns
df.dropna(subset=['duration','description'],inplace = True)

```

```python
#percent of null values
percent_null = df.notnull().sum()*100/df.shape[0]
percent_null
```

```python
#rating coln
df['rating'].unique()
```

```python
df['rating'] = df['rating'].replace({'74 min' : np.nan, '84 min' : np.nan, '66 min': np.nan
                                    , 'TV-Y7-FV' : 'TV-Y7'})
df['rating'].unique()
```

```python
df['year_added'] = df['date_added'].apply(lambda x : x.split(',')[-1])
df['month_added'] =  df['date_added'].apply(lambda x : x.split(',')[0].split()[0])
df.head()
```

```python
#remove unnesecary coln
```

```python
df.drop( [ 'date_added','description','show_id'],axis = 1, inplace = True)
```

```python
df.head()
```

```python
df = df.rename(columns = {'year':'year_added','month':'month_added'})
df.head()
```

```python

genres = df['listed_in'].apply(lambda x:x.split(','))
df['genre_1'] = genres.str.get(0)
df['genre_2'] = genres.str.get(1)
df['genre_3'] = genres.str.get(2)
df.head()
```

```python
## visualisation
sns.countplot(x= 'type',data = df, palette = 'Blues')
plt.show()
```

```python
sns.boxplot(y= df['release_year'])
plt.show()
```

```python
month_added = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
               'August', 'September', 'October', 'November', 'December']
df1 = df.groupby('release_year')['month_added'].value_counts().unstack()[month_added]
```

```python
fig,ax= plt.subplots(1,1,figsize = (10,8))
ax = sns.heatmap(df1,cmap = "magma")
plt.title('Releases')
plt.show()
```

```python
#Top10 director

top_10dir= df[df['director']!="unknown"]['director'].value_counts()[:10].reset_index()
top_10dir
```

```python
sns.barplot(y = top_10dir['director'], x= top_10dir['count'])
labels = list(range(0,25,5))
plt.xticks(ticks = labels)
plt.title("Top 10 directors")
plt.show()
```

```python
country_counts = df['country'].value_counts()
plt.pie(x=country_counts,labels=country_counts.index,autopct='%.1f%%')
plt.show()
```

```python
d = df[df['country'].isin(['United States', 'India', 'United Kingdom'])]
sns.countplot(x='country', data=d)
plt.show()
```

```python
#Ratings
plt.figure(figsize= (15,15))
sns.countplot(x= df['rating'], data = df)
plt.show()
```

```python
pp = df.groupby('type')['release_year'].value_counts().unstack(level= -2)
pp
```
